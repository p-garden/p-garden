# **실습문제 3**

**그리드서치와 랜덤서치로 도미와 농어를 낚으러 가볼까요 ~ ? 🐠🐟**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/sample_data/Fish.csv')

#1. 

fish =  df.loc[(df['Species'] == 'Bream') |(df['Species'] == 'Perch')]    # 
         
data =  np.array(fish[['Weight','Length1', 'Length2','Length3', 'Height', 'Width']])  #
target = fish['Species']  #

# 필요한 생선만 쓰는건 1주차 생선데이트 실습때도 해봤답니당 ~

# Training set and Test set 으로 split 해줍니다 
from sklearn.model_selection import train_test_split
train_input,test_input,train_target,test_target = train_test_split(data, target, test_size = 0.2, random_state = 42)

#Training set 을 Decision Tree Classifier model 에 훈련해줍니다. 
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 42)
dt.fit(train_input, train_target)

# Training set and Test set 에 대한 예측 점수를 확인해줍니다.
print(dt.score(train_input,train_target))
print(dt.score(test_input,test_target))

```

```python
params_gs = {'min_impurity_decrease': np.arange(0.0001,0.001,0.0001),'max_depth': np.arange(5,20,1),
             'min_samples_split': np.arange(2,100,10),'min_samples_leaf' : np.arange(1,10,1) } #
 

# Random Search 모델을 이용하여 적절한 파라미터를 찾아 줍시다!
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(DecisionTreeClassifier(random_state=42),params_gs,n_jobs=-1)
gs.fit(train_input,train_target)

## 적절한 하이퍼 파라미터 값을 찾아봅시다.
print(gs.best_params_)

## 모델이 적절하게 학습되었는지 확인해 봅시다!!
dt = gs.best_estimator_

print(np.max(gs.cv_results_['mean_test_score']))  # 최상의 교차 검증 점수 출력 #
print(dt.score(test_input, test_target)) # test set에 대한 dt.score 출력 #
```

```python
''' 
#4.
랜덤 서치는 매개변수를 균등 분포에서 랜덤 샘플링합니다. 이를 위해 불러와야하는 싸이파이 패키지가 있습니다.
불러와주세요 !
'''
from scipy.stats import uniform,randint #

'''
#5
파라미터의 세가지 값을 채워주세요 ! 

'''
params_rs = {'min_impurity_decrease': np.arange(0.0001,0.001,0.0001),
          'max_depth' : randint(20,50),
         'min_samples_split': randint(2,25),
          'min_samples_leaf':randint(1,25)
          } #

# Random Search 모델을 이용하여 적절한 파라미터를 찾아 줍시다!
from sklearn.model_selection import RandomizedSearchCV
N=100
rs = RandomizedSearchCV(DecisionTreeClassifier(random_state=42),params_rs,n_jobs=-1,n_iter=N,random_state=42)
rs.fit(train_input,train_target)

## 적절한 하이퍼 파라미터 값을 찾아봅시다.
print(rs.best_params_)

## 모델이 적절하게 학습되었는지 확인해 봅시다!!
dt = rs.best_estimator_


print(np.max(rs.cv_results_['mean_test_score']))  # 최상의 교차 검증 점수 출력 #
print(dt.score(test_input, test_target)) # test set에 대한 dt.score 출력 #
```
