# **문제 5**
fish데이터를 이용해 주성분 분석하기 

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#데이터셋 물러오기
fish = pd.read_csv('https://bit.ly/fish_csv_data')
print(fish)
#데이터 전처리
target_species = fish.Species.replace({'Bream':0,'Roach':1, 'Whitefish':2, 'Parkki':3, 'Perch':4 ,'Pike':5, 'Smelt':6})
del fish['Species']
```

🐠 fish데이터에는 5개의 특성이 있습니다. (Weight, Length, Diagonal, Height,Width) 

 따라서, 주성분의 개수에 따른 변화들을 확인해봅시다.

```python
#주성분의 개수를 1부터 5까지 반복하여 각 주성분 마다 분산 확인하기
acc=[]
for i in range(1,6):
  pca = PCA(n_components=i)  
  pca.fit(fish)
  fish_pca= pca.transform(fish)
  acc.append(np.sum(pca.explained_variance_ratio_))
  
print(acc)

#주성분 마다 분산 그래프로 확인하기
plt.plot(pca.explained_variance_ratio_)
plt.show
```

🐠 원본데이터의 성능보다 주성분 분석으로 축소한 데이터의 성능이 좋으려면 어떻게 해야할까요?


```python
lr = LogisticRegression()
target = np.array(target_species)
score_pca=[]

#pca로 축소한 데이터가 원본데이터의 성능을 뛰어넘으려면 n개 이상의 주성분을 분석해야한다. n개 구하기
for i in range(1,6):
  pca = PCA(n_components=i)
  pca.fit(fish)
  fish_pca= pca.transform(fish)
  scores = cross_validate(lr,fish_pca,target)
  score_pca.append(np.mean(scores['test_score']))

scores = cross_validate(lr,fish,target)
original_score = np.mean(scores['test_score'])

```

```python
print(score_pca, original_score)
```

```python
# 그래프로 pca로 축소한 그래프와 원본데이터의 점수 표현하기 (pca는 plot으로 원본은 scatter로 표현(x값에는 정수 2 넣기))
plt.plot(score_pca)
plt.scatter(2,np.mean(scores['test_score']))
plt.show()  
```

🐠 몇개이상의 주성분을 분석해야 원본 데이터의 성능을 뛰어넘었나요? 
-> n:  3개

🐠 이제 몇개이상의 주성분을 분석해야 좋은 성능을 낼 수 있는지도 알았으니
물고기들을 종류별로 산점도로 나타내봅시다.

```python
km = KMeans(n_clusters=7, random_state=42)
km.fit(fish_pca)

#각 물고기 종류별로 클러스터 산점도 표현하기
for label in range(0,7):
  data = fish_pca[km.labels_  == label]
  plt.scatter(data[:,0], data[:,1])

plt.legend([ 'Perch', 'Parkki', 'Smelt', 'Pike', 'Whitefish', 'Roach', 'Bream'])
plt.show()
```
